```python
from selenium import webdriver
from pandas import DataFrame
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import time
```

```python
driver = webdriver.Chrome("chromedriver.exe")
```

```python
def watcha_movie(name):

    driver.get("https://pedia.watcha.com/ko-KR/search")
    sel = "#root > div > div.css-1sh3zvx-NavContainer.ebsyszu0 > header > nav > div > div > ul > li.css-67pwm-NavList.e1cl8ith5 > div > div > form > label > input"
    ui = driver.find_element_by_css_selector(sel)
    ui.send_keys(name)
    ui.send_keys(Keys.RETURN)

    time.sleep(2)

    sel = "#root > div > div.css-1sh3zvx-NavContainer.ebsyszu0 > section > section > section:nth-child(1) > div.css-wyngcf-ScrollBarContainer.e1f5xhlb0 > div.css-chidac-ScrollBar.e1f5xhlb1 > div > div > div > ul > li:nth-child(1) > a"
    ui = driver.find_element_by_css_selector(sel)
    ui.click()

    time.sleep(2)

    sel = "#root > div > div.css-1sh3zvx-NavContainer.ebsyszu0 > section > div > div > div > section > div.css-p3jnjc-Pane.e1svyhwg12 > div > div > div > div > div.css-1xlr4il-ContentRatings.e1svyhwg16"
    ui = driver.find_element_by_css_selector(sel)
    watcha_rate = ui.text
    
    time.sleep(2)
    
    return watcha_rate
```

```python
import pandas as pd

def watch_rate_by_year(year):
    trend_list = pd.read_csv(f"trend/trend_{year}.csv", sep=",", index_col='Unnamed: 0', encoding='cp949')
    list(trend_list['제목'])
    data = [ ] 
    for i in list(trend_list['제목']):
        watcha_rate = watcha_movie(i)
        data.append([i,watcha_rate]) 
    df = DataFrame(data)
    df.columns = ['제목','왓챠평점']
    df.to_csv(f"trend_watcha/trend_{year}_wacha_rate.csv", encoding = 'cp949')
```

```python
for i in range(2012,2021):
    watch_rate_by_year(i)
```

```python
def watch_movie_info(name):
    driver.get("https://pedia.watcha.com/ko-KR/search")
    sel = "#search_input_in_modal_like"
    ui = driver.find_element_by_css_selector(sel)
    ui.send_keys(name)
    ui.send_keys(Keys.RETURN)
    
    time.sleep(2)
    
    sel = "#root > div > div.css-1sh3zvx-NavContainer.ebsyszu0 > section > section > section:nth-child(1) > div.css-wyngcf-ScrollBarContainer.e1f5xhlb0 > div.css-chidac-ScrollBar.e1f5xhlb1 > div > div > div > ul > li:nth-child(1) > a"
    ui = driver.find_element_by_css_selector(sel)
    ui.click()

    time.sleep(2)
    
    sel = "#root > div > div.css-1sh3zvx-NavContainer.ebsyszu0 > section > div > div > div > section > div.css-p3jnjc-Pane.e1svyhwg12 > div > div > div > div > div.css-zv7ww6-Detail.e1svyhwg15"
    ui = driver.find_element_by_css_selector(sel)
    info = ui.text.split(" ・ ")
    
    time.sleep(2)
    
    return info  
```

```python
def watch_info_by_year(year):
    trend_list = pd.read_csv(f"trend/trend_{year}.csv", sep=",", index_col='Unnamed: 0', encoding='cp949')
    list(trend_list['제목'])
    data = [ ] 
    for i in list(trend_list['제목']):
        watcha_info = watch_movie_info(i)
        data.append([i,watcha_info[0],watcha_info[1],watcha_info[2:]]) 
    df = DataFrame(data)
    df.columns = ['제목','연도','장르','제작국가']
    df.to_csv(f"trend_watcha/trend_{year}_info.csv", encoding = 'cp949')
```

```python
for year in range(2013,2021):
    watch_info_by_year(year)
```
