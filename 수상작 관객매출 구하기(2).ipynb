```python
other = pd.read_csv("film_all3.csv", sep=",", encoding='cp949')
other.head()
```

```python
prize_audi = DataFrame()
```

```python
list_tit = list(other['title'])
list_tit = set(list_tit)
list_tit = list(list_tit)
list_tit
```

```python
for i in list_tit:
    try:
        aa = DataFrame(omg.loc[f'{i}'])
        aa = aa.transpose()
        prize_audi = prize_audi.append(aa)
    except:
        pass
```

```python
prize_audi = prize_audi.reset_index()
prize_audi.columns = ['제목', '관객수', '매출액']
prize_audi
```
