```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

info_list = ["title","year","director_names","nations","genres"]
```

### 칸영화제 황금종려상 수상작

```python
Cannes = pd.DataFrame(columns = info_list)
code_list = pd.DataFrame(columns=["code"])
for i in range(1,5):
    url = "https://api-pedia.watcha.com/api/staffmades/609/contents"

    params = {
        "page": i,
        "size": 20
    }

    headers = {
        "origin": "https://pedia.watcha.com",
        "pragma": "no-cache",
        "referer": "https://pedia.watcha.com/",
        "sec-fetch-dest": "empty",
        "sec-fetch-mode": "cors",
        "sec-fetch-site": "same-site",
        "x-watcha-client": "watcha-WebApp",
        "x-watcha-client-language": "ko",
        "x-watcha-client-region": "KR",
        "x-watcha-client-version": "2.0.0",
        "x-watcha-remote-addr": "121.134.241.197",
        "user-agent": "Mozilla/5.0"
    }

    resp = requests.get(url, headers=headers, params=params)
    data = resp.json()
    df = pd.DataFrame(data["result"]["result"])
    Cannes_i = df[info_list]
    Cannes = Cannes.append(Cannes_i)
    #코드 리스트 가져오기
    code_list = code_list.append(df[["code"]])

Cannes = Cannes.reset_index()[info_list]

#코드 리스트로 url 만들기
url_list = []
for i in range(len(code_list)):
    url = "https://pedia.watcha.com/ko-KR/contents/" + code_list.iloc[i]
    url_list.append(url[0])
#url로 별점 가져오기   
rate_list = []
for i in url_list:
    url = i
    resp = requests.get(url)

    soup = BeautifulSoup(resp.text)
    result = soup.select("#root > div > div.css-1sh3zvx-NavContainer.ebsyszu0 > section > div > div > div > section > div.css-p3jnjc-Pane.e1svyhwg12 > div > div > div > div > div.css-1xlr4il-ContentRatings.e1svyhwg16")
    
    for item in result:
        rate_list.append(item.text)           

Cannes["rate"]= rate_list
Cannes
```

```python
Cannes["rate"]= rate_list
Cannes
```

```python
#to_csv

Cannes.to_csv("Cannes.csv", encoding="cp949")
```
