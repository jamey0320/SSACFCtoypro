```python
import pandas as pd
from pandas import DataFrame
import time
```

```python
flim_all2 = pd.read_csv("flim_all2.csv", sep=",", index_col='Unnamed: 0', encoding='cp949')
flim_all2
```

```python
flim_all2["rate"]=flim_all2["rate"].str[0:3]
```

```python
flim_all2.columns=['연도', '제목', '감독', '제작국가', '장르', '왓챠평점', '영화제']
```

```python
film_all2 = DataFrame()
```

```python
len(flim_all2['연도'])
```

```python
flim_all2['연도'] = pd.to_numeric(flim_all2['연도'])
for i in range(len(flim_all2['연도'])):
    if flim_all2['연도'][i] >= 2010:
        film_all2 = film_all2.append(flim_all2.iloc[i])
```

```python
film_all2
```

```python
film_all2 = film_all2.astype({"연도": 'string'})
film_all2["연도"]=film_all2["연도"].str[0:4]
film_all2=film_all2[['연도', '제목', '감독', '제작국가', '장르', '왓챠평점', '영화제']]
film_all2
```

```python
film_all2.to_csv('film_all2.csv', encoding="cp949")
```
