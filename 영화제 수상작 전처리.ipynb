```python
import pandas as pd
```

```python
flim_all = pd.read_csv("flim_all.csv", encoding='cp949')
flim_all
```

```python
for i in range(len(flim_all["director_names"])):
    flim_all["director_names"][i] = flim_all["director_names"][i][1:-1]
    flim_all["director_names"][i] = flim_all["director_names"][i].replace("'","")
```

```python
for i in range(len(flim_all["genres"])):
    flim_all["genres"][i] = flim_all["genres"][i][1:-1]
    flim_all["genres"][i] = flim_all["genres"][i].replace("'","")
```

```python
for i in range(len(flim_all["nations"])):
    flim_all["nations"][i] = flim_all["nations"][i][1:-1]
    flim_all["nations"][i] = flim_all["nations"][i].replace("{","")
    flim_all["nations"][i] = flim_all["nations"][i].replace("}","")
    flim_all["nations"][i] = flim_all["nations"][i].replace("'name': ","")
    flim_all["nations"][i] = flim_all["nations"][i].replace("'","")
```

```python
for i in range(len(flim_all["rate"])):
    flim_all["rate"][i] = flim_all["rate"][i].replace("평균 ★","")
```
